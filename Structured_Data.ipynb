```python
import pandas as pd
import ast
from pandas import json_normalize

# Function to normalize a column with JSON-like string data
def normalize(df, column_name):
    df[f"{column_name}_dict"] = df[column_name].apply(ast.literal_eval)
    return json_normalize(df[f"{column_name}_dict"])

# Function to extract key-value pairs from car overview data
def extract_key_value_dicts(details):
    if isinstance(details, list):
        return {df['key']: df['value'] for df in details}
    return {}

# Function to process car overview column
def process_car_overview(car_overview_df):
    car_overview_df.drop(columns=["bottomData"], inplace=True, axis=1)
    new_car_overview = car_overview_df['top'].apply(extract_key_value_dicts)
    return pd.json_normalize(new_car_overview)

# Function to extract key-value pairs from car features data
def extract_key_value_dicts1(details_1):
    if isinstance(details_1, list):
        return {i: df['value'] for i, df in enumerate(details_1) if 'value' in df}
    return {}


# Function to extract key-value pairs from car specs data
def extract_key_value_dicts2(details2):
    if isinstance(details2, list):
        return {df['key']: df['value'] for df in details2}
    return {}

# Function to process car specs column
def process_car_specs(car_specs_df):
    car_specs_df.drop(columns=["commonIcon"], inplace=True, axis=1)
    new_car_specs_top = car_specs_df["top"].apply(extract_key_value_dicts2)
    return pd.json_normalize(new_car_specs_top)

# Function to process specs data by index
def process_car_specs_data_by_index(new_car_specs_data, index, extract_func):
    specs_data = pd.json_normalize(new_car_specs_data[index])
    if 'list' in specs_data.columns:
        extracted_data = specs_data["list"].apply(extract_func)
        return pd.json_normalize(extracted_data)
    return pd.DataFrame()

# Main function to process car data for a specific city
def process_city_data(file_path, city_name):
    df_city = pd.read_excel(file_path)
    df_city.drop(columns=["car_links"], inplace=True, axis=1)

    # Process each column
    new_car_detail = normalize(df_city, "new_car_detail")
    new_car_overview = process_car_overview(normalize(df_city, "new_car_overview"))

    # Process specs column and data
    new_car_specs = normalize(df_city, "new_car_specs")
    new_car_specs_top = process_car_specs(new_car_specs)
    new_car_specs_data = pd.json_normalize(new_car_specs["data"])

    specs_list_1 = process_car_specs_data_by_index(new_car_specs_data, 0, extract_key_value_dicts2)
    specs_list_2 = process_car_specs_data_by_index(new_car_specs_data, 1, extract_key_value_dicts2)
    specs_list_3 = process_car_specs_data_by_index(new_car_specs_data, 2, extract_key_value_dicts2)

    # Combine data into final DataFrame
    df_city_cars = pd.concat([
        new_car_detail, new_car_overview, new_car_specs_top,
        specs_list_1, specs_list_2, specs_list_3
    ], axis=1)
    df_city_cars['City'] = city_name

    return df_city_cars

# Process data for bangalore_cars
df_bangalore_cars = process_city_data("bangalore_cars.xlsx", "bangalore")
df_bangalore_cars
```

```python
import pandas as pd
import ast
from pandas import json_normalize

# Function to normalize a column with JSON-like string data
def normalize(df, column_name):
    df[f"{column_name}_dict"] = df[column_name].apply(ast.literal_eval)
    return json_normalize(df[f"{column_name}_dict"])

# Function to extract key-value pairs from car overview data
def extract_key_value_dicts(details):
    if isinstance(details, list):
        return {df['key']: df['value'] for df in details}
    return {}

# Function to process car overview column
def process_car_overview(car_overview_df):
    car_overview_df.drop(columns=["bottomData"], inplace=True, axis=1)
    new_car_overview = car_overview_df['top'].apply(extract_key_value_dicts)
    return pd.json_normalize(new_car_overview)

# Function to extract key-value pairs from car features data
def extract_key_value_dicts1(details_1):
    if isinstance(details_1, list):
        return {i: df['value'] for i, df in enumerate(details_1) if 'value' in df}
    return {}


# Function to extract key-value pairs from car specs data
def extract_key_value_dicts2(details2):
    if isinstance(details2, list):
        return {df['key']: df['value'] for df in details2}
    return {}

# Function to process car specs column
def process_car_specs(car_specs_df):
    car_specs_df.drop(columns=["commonIcon"], inplace=True, axis=1)
    new_car_specs_top = car_specs_df["top"].apply(extract_key_value_dicts2)
    return pd.json_normalize(new_car_specs_top)

# Function to process specs data by index
def process_car_specs_data_by_index(new_car_specs_data, index, extract_func):
    specs_data = pd.json_normalize(new_car_specs_data[index])
    if 'list' in specs_data.columns:
        extracted_data = specs_data["list"].apply(extract_func)
        return pd.json_normalize(extracted_data)
    return pd.DataFrame()

# Main function to process car data for a specific city
def process_city_data(file_path, city_name):
    df_city = pd.read_excel(file_path)
    df_city.drop(columns=["car_links"], inplace=True, axis=1)

    # Process each column
    new_car_detail = normalize(df_city, "new_car_detail")
    new_car_overview = process_car_overview(normalize(df_city, "new_car_overview"))

    # Process specs column and data
    new_car_specs = normalize(df_city, "new_car_specs")
    new_car_specs_top = process_car_specs(new_car_specs)
    new_car_specs_data = pd.json_normalize(new_car_specs["data"])

    specs_list_1 = process_car_specs_data_by_index(new_car_specs_data, 0, extract_key_value_dicts2)
    specs_list_2 = process_car_specs_data_by_index(new_car_specs_data, 1, extract_key_value_dicts2)
    specs_list_3 = process_car_specs_data_by_index(new_car_specs_data, 2, extract_key_value_dicts2)

    # Combine data into final DataFrame
    df_city_cars = pd.concat([
        new_car_detail, new_car_overview, new_car_specs_top,
        specs_list_1, specs_list_2, specs_list_3
    ], axis=1)
    df_city_cars['City'] = city_name

    return df_city_cars

# Process data for jaipur_cars
df_jaipur_cars = process_city_data("jaipur_cars.xlsx", "jaipur")
df_jaipur_cars
```

```python
# List of DataFrames
all_cities_car_data = [df_banglore_cars,df_chennai_cars,df_delhi_cars,df_hyderabad_cars,df_jaipur_cars,df_kolkata_cars]

# 1. Rename Duplicate Columns to Ensure Unique Column Names
def rename_duplicate_columns(df):
    # Find duplicate columns
    cols = pd.Series(df.columns)
    for dup in cols[cols.duplicated()].unique():
        # Rename duplicates by appending a suffix '_1', '_2', etc.
        count = 0
        for i in range(len(cols)):
            if cols[i] == dup:
                count += 1
                if count > 1:
                    cols[i] = f"{dup}_{count-1}"
    df.columns = cols
    return df

# Apply the function to each DataFrame to ensure unique column names
all_cities_car_data = [rename_duplicate_columns(df) for df in all_cities_car_data]

# 2. Ensure All DataFrames Have the Same Columns
# Get the set of all unique column names across all DataFrames
all_columns = set()
for df in all_cities_car_data:
    all_columns.update(df.columns)

# Add missing columns to each DataFrame with NaN values
for df in all_cities_car_data:
    missing_cols = all_columns - set(df.columns)
    for col in missing_cols:
        df[col] = pd.NA  # or use np.nan if you prefer

# 3. Concatenate DataFrames with Outer Join
# Reset index for each DataFrame to avoid index conflicts
all_cities_car_data = [df.reset_index(drop=True) for df in all_cities_car_data]

# Concatenate all DataFrames along rows (axis=0) with outer join
combined_df = pd.concat(all_cities_car_data, axis=0, join='outer', ignore_index=True)

print(combined_df.shape)
combined_df.head()
```
